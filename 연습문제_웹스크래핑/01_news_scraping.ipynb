```python
# reqeusts, bs4 import
import requests
import bs4
# BeautifulSoup 클래스 import
from bs4 import BeautifulSoup
```

```python
# 기후/환경 뉴스 
#url = 'https://news.daum.net/climate'

#dict타입으로 요청 파라미터 설정
req_param = {
    'sid': 'climate'
}
# 
url = 'https://news.daum.net/{sid}   '.format(**req_param)
print(url)

# 요청 헤더 설정 : 브라우저 정보 (사람처럼 보이게 하기 위함)
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/144.0.0.0 Safari/537.36'
}

# requests 의 get() 함수 호출하기 
res = requests.get(url, headers=req_header)
print(res.status_code)
print(res.ok)
print(type(res))
#print(res.text)
# 응답(response)이 OK 이면
# 응답 (response)에서 text 추출
# BeautifulSoup 객체 생성  
if res.ok:
    soup = BeautifulSoup(res.text,'html.parser')
    print(len(soup.select("ul.la_text a[href*='https://t1.daumcdn.net/']")))
    # CSS 선택자를 사용해서 a tag 목록 가져오기
    a_tags = soup.select("ul.la_text a[href*='https://t1.daumcdn.net/']")
    print(type(a_tags), type(a_tags[0])) #[Tag,Tag]
    # <a> 태그 리스트 순회하기 
    for a_tag in a_tags:
        title = a_tag.text.strip()
        link = a_tag['href']
        print(title, link)

else:
    print(f'Error Code = {res.status_code}')

# 응답(response)이 Error 이면 status code 출력    

```

```python
import requests
from bs4 import BeautifulSoup

# 1. 요청 파라미터 설정 (경제 섹션)
req_param = {
    'sid': 'economy'
}

# 2. URL 설정 (공백 제거 후 출력)
url = 'https://news.daum.net/{sid}'.format(**req_param).strip()
print(url)

# 3. 요청 헤더 설정 (다음 pcweb 환경 대응)
req_header = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/133.0.0.0 Safari/537.36',
    'referer': 'https://news.daum.net/'
}

# 4. requests의 get() 함수 호출 및 정보 출력
res = requests.get(url, headers=req_header)
res.encoding = 'utf-8' # 한글 깨짐 방지

print(type(res))          # <class 'requests.models.Response'> 출력
print(res.status_code)    # 200 출력

# 5. 응답이 OK일 경우 데이터 추출
if res.ok:
    soup = BeautifulSoup(res.text, 'html.parser')
    
    # [변경] ResultSet 개수 확인을 위해 다음 뉴스 헤드라인 리스트 추출
    a_tags = soup.select('ul.list_newsheadline2 li')
    
    # <class 'bs4.element.ResultSet'> 개수 출력
    print(f"{type(a_tags)} {len(a_tags)}")

    # 6. 리스트 순회하며 링크와 제목 출력
    for li in a_tags:
        try:
            # (1) a 태그에서 링크(href) 추출
            a_tag = li.find('a')
            if not a_tag: continue
            link = a_tag['href']
            
            # (2) 제목 추출 (div.cont_thumb 내부의 strong.tit_txt)
            # 다음 뉴스 최신 구조인 'cont_thumb' 클래스를 사용합니다.
            strong_tag = li.select_one('div.cont_thumb strong.tit_txt')
            
            if strong_tag:
                title = strong_tag.text.strip()
                
                print(title, link)

            else:
                print(f'Error Code = {res.status_code}')
```
