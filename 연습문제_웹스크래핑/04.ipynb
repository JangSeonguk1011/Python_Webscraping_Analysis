```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['font.family'] = 'Malgun Gothic'

data = pd.read_csv('data/인구현황.csv')
data = data[data['행정기관'] != '전국']

def gender_ratio(ratio):
    if ratio > 1.0:
        return '남초'
    elif ratio < 1.0:
        return '여초'
    else:
        return '동일'

data['남초여초'] = data['남여 비율'].apply(gender_ratio)
result = data.sort_values(by='세대수', ascending=False)

figure, (ax1, ax2) = plt.subplots(nrows=2, ncols=1, figsize=(18, 12))

sns.countplot(data=data, x='남초여초', ax=ax1, hue='남초여초', legend=False)
ax1.set_title('남초 vs 여초 지역 개수', fontsize=18)
sns.barplot(data=result, x='행정기관', y='세대수', ax=ax2, hue='행정기관', palette='husl', legend=False)
ax2.set_title('지역별 세대수', fontsize=18)

plt.tight_layout()

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 설정
plt.rcParams['font.family'] = 'Malgun Gothic'

# 1. 데이터 로드 및 전처리
data = pd.read_csv('data/인구현황.csv')
data = data[data['행정기관'] != '전국'] # 전국 데이터 제외

# 사진처럼 남자 vs 여자 인구수를 나란히 그리기 위해 데이터 재구조화 (Melt)
# 행정기관별로 '남자 인구수'와 '여자 인구수'를 하나의 열로 모읍니다.
df_melted = data.melt(id_vars=['행정기관'], 
                      value_vars=['남자 인구수', '여자 인구수'], 
                      var_name='성별', value_name='인구수')

# 2. 그래프 생성
figure, ax = plt.subplots(figsize=(18, 8))

# 3. 그룹형 막대 그래프 그리기
sns.barplot(data=df_melted, x='행정기관', y='인구수', hue='성별', ax=ax)

# 4. 마무리 설정
ax.set_title('남자 vs 여자 인구 비교', fontsize=18)
ax.set_xlabel('행정기관', fontsize=12)
ax.set_ylabel('인구수', fontsize=12)

# 범례(Legend) 위치 설정
ax.legend(title='성별')

# x축 행정기관 이름 회전
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
